```python
from torch.utils.data import DataLoader
import pytorch_lightning as pl
from dataset import Dataset
from models import Model
```

```python
train_ds = Dataset(
    dataset="cmapss/FD001",
    split="train",
    normalization="min-max",
)
train_dl = DataLoader(
    train_ds,
    shuffle=True,
    num_workers=2,
    batch_size=len(train_ds),
    pin_memory=True,
    persistent_workers=True,
)
input_size = (train_ds.num_channels, train_ds.window, train_ds.num_features)
input_size
```

```python
model = Model(input_size)
trainer = pl.Trainer(
    gpus=[0],
    max_epochs=10000,
    log_every_n_steps=100,
    logger=pl.loggers.TensorBoardLogger(".", "lightning_logs"),
    auto_scale_batch_size=True,
)
trainer.fit(model, train_dl)
```

```python
test_ds = Dataset(
    dataset="FD001",
    split="test",
    normalization="min-max",
)
test_dl = DataLoader(
    test_ds,
    shuffle=False,
    num_workers=2,
    batch_size=1,
    pin_memory=True,
    persistent_workers=True,
)
input_size = (test_ds.num_channels, test_ds.window, test_ds.num_features)

trainer = pl.Trainer(default_root_dir=".")

model = Model.load_from_checkpoint(
    "lightning_logs/version_0/checkpoints/epoch=9999-step=9999.ckpt",
    input_size=input_size,
    model="linear",
)

trainer = pl.Trainer(default_root_dir=".")
trainer.test(model, test_dataloaders=test_dl, verbose=False)
```
